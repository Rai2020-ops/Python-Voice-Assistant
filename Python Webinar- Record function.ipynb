### Speech Recognition

```python
#Speech to Text Conversion
duration=int(4)
#!pip install SpeechRecognition [Command to install SpeechRecognition package]
import speech_recognition as sr #Library for performing speech recognition with the Google Speech Recognition API.
r=sr.Recognizer()
with sr.Microphone() as source: # use the default microphone as the audio source
    print('Please Speak')
    r.pause_threshold=1 # pausing time
    audio=r.record(source,duration=duration)
    try:  # recognize speech using Google Speech Recognition
        text=r.recognize_google(audio,language='en-In').lower()
        print('\n'+text+'\n')
    except sr.UnknownValueError:           
        exit()
        
#NOTE: this requires PyAudio because it uses the Microphone class
```

```python
#Speech to Text Conversion in hindi

#!pip install SpeechRecognition [Command to install SpeechRecognition package]
import speech_recognition as sr #Library for performing speech recognition with the Google Speech Recognition API.
r=sr.Recognizer()
with sr.Microphone() as source: # use the default microphone as the audio source
    print('Please Speak')
    r.pause_threshold=1 # pausing time
    r.adjust_for_ambient_noise(source,duration=1) #listen for 1 second to calibrate the energy threshold for ambient noise levels
    audio=r.record(source,duration=duration)  # now when we listen, the energy threshold is already set to a good value, and we can reliably catch speech right away
    try:  # recognize speech using Google Speech Recognition
        text=r.recognize_google(audio,language='hi-In').lower()
        print('\n'+text+'\n')
    except sr.UnknownValueError:           
        exit()
        
#NOTE: this requires PyAudio because it uses the Microphone class
```
